To train individual client models, four machine learning algorithms were initially selected, each offering unique strengths:

1. **Logistic Regression:** A linear model for binary classification, providing probabilities and assigning the class with the highest probability.

2. **Support Vector Machine (SVM):** A powerful classification algorithm effective for linear and non-linear data, aiming to find optimal hyperplanes.

3. **Random Forest:** An ensemble method building decision trees and merging them for stable predictions in both classification and regression tasks.

4. **Feedforward Neural Networks (FNN):** A multilayer perceptron capable of learning complex patterns and relationships, suitable for various tasks, including classification.

The evaluation process incorporates standard metrics such as accuracy, precision, recall, and F1 score. It assesses both individual client models and the collaborative global model's effectiveness in addressing cybersecurity challenges.


```python
# IMPORT NECESSARY LIBRARIES
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# READ BOTH DATASET
dataset_1 = pd.read_csv('Dataset/dataset_1.csv')
dataset_2 = pd.read_csv('Dataset/dataset_2.csv')
pd.set_option('display.max_columns',None)
```

Data Preprocessing

```python
# CHECK MISSING VALUES
missing_values = dataset_1.isna().sum()
for m in missing_values:
    if m!=0:
        print(m)
```

```python
missing_values = dataset_2.isna().sum()
for m in missing_values:
    if m!=0:
        print(m)
```

```python
dataset_1['Label'].value_counts()
```

```python
dataset_2['Label'].value_counts()
```

```python
# REMOVING WHITE SPACE FROM BOTH DATASET CLASS COLUMN
dataset_1['Label'] = dataset_1['Label'].str.strip()
dataset_2['Label'] = dataset_2['Label'].str.strip()
```

```python
# CONVERTING CATEGORICAL CLASS INTO NUMERICAL VALUE
mapping_1 = {'DDoS': 1, 'BENIGN': 2}
dataset_1['Label'] = dataset_1['Label'].map(mapping_1)
dataset_1
```

```python
mapping_2 = {'PortScan': 1, 'BENIGN': 2}
dataset_2['Label'] = dataset_2['Label'].map(mapping_2)
dataset_2
```

Model Training and Evaluation

```python
def train_model_and_print_report(X, y, clf):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1234)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    cr = classification_report(y_test, y_pred)
    print(cr)
```

```python
# CREATE INSTANCES OF MACHINE LEARNING MODELS
lr = LogisticRegression(max_iter=1000, random_state=1234)
svc = SVC(kernel='rbf', C=1, gamma='scale', random_state=1234)
rf = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, random_state=1234)
fnn = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=1234)
```

Run These Models For Dataset-1

```python
# DEFINE X & y
X1 = dataset_1.drop(['Label','IP'],axis=1)
y1 = dataset_1['Label']
```

```python
train_model_and_print_report(X1, y1, lr)
```

```python
train_model_and_print_report(X1, y1, svc)
```

```python
train_model_and_print_report(X1, y1, rf)
```

```python
train_model_and_print_report(X1, y1, fnn)
```

# Decision: Random Forest emerged as the most effective model for Dataset-1, achieving perfect scores across accuracy, precision, recall, and F1-score, all registering at 1. Given its exceptional performance, Random Forest stands out as the optimal choice for Dataset-1.

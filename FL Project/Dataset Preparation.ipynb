```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
client_1 = pd.read_csv('MachineLearningCVE/client_1.csv')
client_1.head(10)
```

```python
client_1.shape
```

```python
client_1.columns
```

```python
# REMOVING UNNECESSARY WHITE SPACES
client_1.columns = client_1.columns.str.strip()
client_1.columns
```

```python
client_1['Label'].value_counts()
```

Given the dataset's extensive size and imbalanced class distribution, I have decided to retain 500 data objects from each class. Consequently, the new shape of the dataset will be (1000, 79)

```python
# RANDOMLY SELECTING 500 DATA OBJECT FROM DDoS CLASS
DDoS_filter = client_1['Label'] == 'DDoS'
DDoS_df = client_1[DDoS_filter].sample(n=500, random_state=1234)
DDoS_df
```

```python
# RANDOMLY SELECTING 500 DATA OBJECT FROM BENIGN CLASS
BENIGN_filter = client_1['Label'] == 'BENIGN'
BENIGN_df = client_1[BENIGN_filter].sample(n=500, random_state=1234)
BENIGN_df
```

```python
# MERGING TWO CLASS
merged_df = pd.concat([DDoS_df, BENIGN_df], ignore_index=True)
merged_df
```

```python
# SAVING THE FIRST DATASET
merged_df.to_csv('Dataset/dataset_1.csv')
```

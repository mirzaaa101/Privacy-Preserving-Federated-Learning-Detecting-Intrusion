```python
import pickle
from matplotlib import pyplot as plt

nohomo = True
```

```python
with open(f"LogisticRegression/result-1-{nohomo}.pkl", "rb") as f:
    result1 = pickle.load(f)
with open(f"LogisticRegression/result-2-{nohomo}.pkl", "rb") as f:
    result2 = pickle.load(f)

plt.plot(result1["accuracy"], label="client 1")
plt.plot(result2["accuracy"], label="client 2")
plt.title("LogisticRegression Evaluation Accuracy")
plt.legend()
plt.show()
```

```python
with open(f"MLP/result-1-{nohomo}.pkl", "rb") as f:
    result1 = pickle.load(f)
with open(f"MLP/result-2-{nohomo}.pkl", "rb") as f:
    result2 = pickle.load(f)

plt.plot(result1["accuracy"], label="client 1")
plt.plot(result2["accuracy"], label="client 2")
plt.title("MLP Evaluation Accuracy")
plt.legend()
plt.show()

plt.plot(result1["loss"], label="client 1")
plt.plot(result2["loss"], label="client 2")
plt.title("MLP Evaluation Loss")
plt.legend()
plt.show()
```
